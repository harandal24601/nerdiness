## Correlation Analysis

```python
question_train = train.iloc[:, 0:26]
question_train
```

```python
corr_train = question_train.corr()
corr_train = corr_train.apply(lambda x : round(x, 2))
corr_train
```

```python
corr_train['Q1'].sort_values(ascending=False)
```

```python
unstack_train = corr_train.unstack()
unstack_train
```

```python
df = pd.DataFrame(unstack_train[unstack_train < 1].sort_values(ascending=False), columns=['corr'])
df.style.background_gradient(cmap='viridis')
```
